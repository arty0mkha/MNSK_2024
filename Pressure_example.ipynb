```python
import numpy as np
import matplotlib.pyplot as plt
from Neumann_Pressure import pure_Neumann_Poisson_solver
```

```python

Nx = 25
Ny = 25
L = 1
H = 1
mu = 0.1
w = 0.001
v_in = 0.1
Q = v_in*mu/(w**2)
A = w**3/mu*np.ones(shape=(Ny, Nx+2))
A.T[0] = A.T[-1]*10
f = np.zeros(shape=(Ny,Nx))
top = np.zeros(Nx) 
bottom= np.zeros(Nx) 
left = -Q*np.ones(Ny) 
right = Q*np.ones(Ny) 
```

```python
P, error = pure_Neumann_Poisson_solver(A, f, top, bottom, left, right, Nx ,Ny, L, H, True)
plt.imshow(P, extent=[-L/Nx, L+L/Nx, -H/Ny, H+H/Ny])
plt.colorbar()
error # %
```
